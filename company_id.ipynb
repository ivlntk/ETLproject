```python
matched_companies = []

for row in ct_output_co.index:
    company_name = ct_output_co.get_value(row,"Company")
    for columns in company_id.index:
        company_id_name=company_id.get_value(columns,"Name")
        matched_token=fuzz.partial_ratio(company_name,company_id_name)
        if matched_token >99:
            matched_companies.append([company_name,company_id_name,matched_token])
```

```python
lists = matched_companies
lists
```
